# Pacotes usados

```python
from Bio import Seq, SeqIO
```

# Código para importação e contagem de nucleotídios


```python
data_path = "/content/drive/MyDrive/AB/exercicio_2/data/Dengue virus type 1 strain Den1BR-90 sequence.fasta"
```

# Dois modos de carregamento dos dados FASTA

```python
data_1 = SeqIO.read(data_path, "fasta")
```

```python
data_2 = open(data_path, 'r').read()
```

# Dois modos de resolução do problema

```python
A_1 = data_1.count('A')
T_1 = data_1.count('T')
C_1 = data_1.count('C')
G_1 = data_1.count('G')

total_1 = A_1 + T_1 + C_1 + G_1
```

```python
A_2 = data_2.count('A')
T_2 = data_2.count('T')
C_2 = data_2.count('C')
G_2 = data_2.count('G')

total_2 = A_2 + T_2 + C_2 + G_2
```

# Criação dos arquivos

```python
output_1 = f"A: {A_1} T: {T_1} C: {C_1} G: {G_1} Total: {total_1}"

output_2 = f"A: {A_2} T: {T_2} C: {C_2} G: {G_2} Total: {total_2}"
```

```python
file_1 = open("/content/drive/MyDrive/AB/exercicio_2/arquivo_metodo_comum.txt", 'w')
file_2 = open("/content/drive/MyDrive/AB/exercicio_2/arquivo_metodo_fasta.txt", 'w')

file_1.write(output_1)
file_2.write(output_2)

file_1.close()
file_2.close()
```
