```python
def ATCGCounter():
  A = 0
  T = 0
  C = 0
  G = 0

  DNA = input("Dê-me uma string DNA\n")

  for i in range(len(DNA)):
    if DNA[i] == 'A':
      A = A + 1
    elif DNA[i] == 'T':
      T = T + 1
    elif DNA[i] == 'C':
      C = C + 1
    elif DNA[i] == 'G':
      G = G + 1

  totalSum = A + T + C + G

  return (A, T, C, G, totalSum)
```

```python
result = ATCGCounter()
```

```python
print(f"A: {result[0]} \nT: {result[1]} \nC: {result[2]} \nG: {result[3]} \nTotal: {result[4]} nucleotídeos")
```
